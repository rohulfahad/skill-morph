```python
# Task 1: Basic Matplotlib (3 plots)
import pandas as pd
#file_path = '/content/drive/My Drive/gym.csv'
df = pd.read_csv("/content/drive/My Drive/gym.csv")
import matplotlib.pyplot as plt

# 1. Bar chart of gym membership types
df['abonoment_type'].value_counts().plot(kind='bar', color='red')
plt.title("Gym Membership Types")
plt.xlabel("Membership Type")
plt.ylabel("Count")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 2. Histogram of time in gym
plt.hist(df['avg_time_in_gym'],bins=10,color='orange',edgecolor='black')
plt.title("Histogram of Average Time in Gym")
plt.xlabel("Time (minutes)")
plt.ylabel("Frequency")
plt.show()

# 3. Scatter plot: Age vs visits per week
plt.scatter(df['Age'], df['visit_per_week'], color='pink', alpha=0.7)
plt.title("Age vs Visits per Week")
plt.xlabel("Age")
plt.ylabel("Visits per Week")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Box plot: Gender vs Age
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='gender', y='Age')
plt.title("Box Plot: Gender vs Age")
plt.show()


# 2. Count plot: Days per week
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='days_per_week', palette='Set2')
plt.title("Count Plot: Days per Week")
plt.show()

# 3. Bar plot: Average visits by membership type
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='abonoment_type', y='visit_per_week', palette='pastel')
plt.title("Average Visits by Membership Type")
plt.show()
```
