```python
#Load the Data
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# TODO: Load the housing data
df = pd.read_csv('/content/Housing.csv')

# TODO: How many houses are there?
print("Number of houses:", len(df))


# TODO: Show first 3 houses
print("First 3 houses:")
# Your code here
print(df.head(3))
```

```python
from sklearn.preprocessing import LabelEncoder
#Convert Yes/No to 1/0
yes_no_columns=['mainroad',"guestroom","basement","hotwaterheating","airconditioning","parking","prefarea","furnishingstatus"]
df[yes_no_columns]=df[yes_no_columns].replace({"yes":1,"no":0})
print(df[yes_no_columns])

# Check what's in our data
print(df.head())

print("Current mainroad values:", df['mainroad'].head())
# Shows: yes, no, yes, yes, no
yes_no_cols=['mainroad',"guestroom","basement","hotwaterheating","airconditioning","parking","prefarea","furnishingstatus"]
for col in yes_no_cols:
    print(df[col].all)
# Models need numbers, not text!
print("_ _"*40)
le=LabelEncoder()
df['furnishingstatus']=le.fit_transform(df['furnishingstatus'])
print(df['furnishingstatus'].all)

# We need: yes → 1, no → 0
print(df.info)

```

```python
# Split into Train and Test
df=df.drop('area',axis=1)

df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Train-test split (80% train, 20% test)
train_size = int(0.8 * len(df))

train_df = df.iloc[:train_size]
test_df = df.iloc[train_size:]

print("Train Set:")
print(train_df)
print("\nTest Set:")
print(test_df)

# TODO: Split - 80% train, 20% test
```
